```python
import requests
from bs4 import BeautifulSoup

url = "https://www.ptt.cc/bbs/IA/index.html"

for times in range(20):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    articles = soup.select('.title a')
    paging = soup.select('.btn-group.btn-group-paging a')
    
    last_url = "https://www.ptt.cc" + paging[1]['href']

    for each_title in articles:
        title.append(each_title.text)
        print(each_title.text)
        

    
    url = last_url
    
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] #提供字型避免圖軸標題中文產生亂碼

colrogroup1 = ['#427f8f','#4a8fa1','#559db0','#66a7b8','#77b1c0','#89bbc8','#9ac5d0','#bdd9e0','#cee3e8','#e0edf0']
#藍色漸層
colrogroup2 = ['#cd0003','#e60003','#ff0004','#ff1a1d','#ff3436','#ff4d4f','#ff6768','#ff8181','#ff9a9b','#ffb4b4']
#紅色漸層

news_brand = ['美國','韓國','中國','日本','越南',
         '英國','香港','新加坡','加拿大','印度', '巴西', '阿根廷', '法國']


brand_count = []

for name in news_brand:
    count = 0
    for mentioned in title:
        if name in str(mentioned):
            count = count +1
    brand_count.append(count)
    
    
    
brand_count
```

```python
plt.bar(news_brand, brand_count, color=colrogroup1) 
plt.xticks(fontsize=15,rotation=90)
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

for index, data in enumerate(brand_count):
    plt.text(x=index, y =data, s=data, verticalalignment="top", horizontalalignment="center")
    
plt.xlabel('新聞標題名稱', fontsize=15)
plt.ylabel('聲量', fontsize=15)
plt.title('ptt標題國家次數', fontsize=20)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] #提供字型避免圖軸標題中文產生亂碼

colrogroup1 = ['#427f8f','#4a8fa1','#559db0','#66a7b8','#77b1c0','#89bbc8','#9ac5d0','#bdd9e0','#cee3e8','#e0edf0']
#藍色漸層
colrogroup2 = ['#cd0003','#e60003','#ff0004','#ff1a1d','#ff3436','#ff4d4f','#ff6768','#ff8181','#ff9a9b','#ffb4b4']
#紅色漸層

#決定所欲查詢比對之聲量標的
country_brand = ['美國','韓國','中國','日本','越南',
         '英國','香港','新加坡','加拿大','印度', '巴西', '阿根廷', '法國']


country_count = []

for name in country_brand:
    count = 0
    for mentioned in title:
        if name in str(mentioned):
            count = count +1
    country_count.append(count)
    

country_count
```

```python
plt.bar(country_brand, country_count, color=colrogroup1) 
plt.xticks(fontsize=15,rotation=90)
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

for index, data in enumerate(country_count):
    plt.text(x=index, y =data, s=data, verticalalignment="top", horizontalalignment="center")
    
plt.xlabel('新聞標題名稱', fontsize=15)
plt.ylabel('聲量', fontsize=15)
plt.title('新聞標題名稱聲量分析', fontsize=20)
plt.show()

```

```python
# import matplotlib相關套件
import matplotlib.pyplot as plt
# import字型管理套件
from matplotlib.font_manager import FontProperties
 
 

 #設定圖片大小為長15、寬10
plt.figure(figsize=(15,10),linewidth = 2)
# 把資料放進來並指定對應的X軸、Y軸的資料，用方形做標記(s-)，並指定線條顏色為紅色，使用label標記線條含意
plt.plot(country_brand,country_count,'s-',color = 'r', label="工商時報")
# 把資料放進來並指定對應的X軸、Y軸的資料 用圓形做標記(o-)，並指定線條顏色為綠色、使用label標記線條含意
plt.plot(news_brand,brand_count,'o-',color = 'g', label="PTT")

plt.title("標題國家折線圖", x=0.5, y=1.03)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
# 標示x軸(labelpad代表與圖片的距離)
plt.xlabel("國家名稱", fontsize=30, labelpad = 15)
# 標示y軸(labelpad代表與圖片的距離)
plt.ylabel("標題出現字數", fontsize=30, labelpad = 20)
# 顯示出線條標記位置
plt.legend(loc = "best", fontsize=20)
# 畫出圖片
plt.show()
```
