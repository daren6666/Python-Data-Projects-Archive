```python
import requests, json
import matplotlib.pyplot as plt

url = "https://api.hahow.in/api/group/music/courses?page=1"


res = requests.get(url,headers = {"user-agent":"XXX"})

Json_data = json.loads(res.text)

a = Json_data['data']

c = []
d = []

for b in a:
    if (b['status']=="SUCCESS"):
        continue
    c.append(b['numRating'])
    d.append(b['numSoldTickets'])
    print("價格",b['basePricingInfo']['price'],"報名人數",b['numSoldTickets'],"評價次數",b['numRating'])
print(c)
print(d)


```

```python
p=[]
m=[]
for b in a:
    if (b['status']=="SUCCESS"):
        continue
    if(b['basePricingInfo']['price']>=2500):
        p.append(b['numRating'])
        m.append(b['numSoldTickets'])
        print("價格",b['basePricingInfo']['price'],"報名人數",b['numSoldTickets'],"評價次數",b['numRating'])
print(p)
print(m)
```

```python
q=[]
w=[]
for b in a:
    if (b['status']=="SUCCESS"):
        continue
    if(b['basePricingInfo']['price']>=1500 and b['basePricingInfo']['price']<=2500):
        q.append(b['numRating'])
        w.append(b['numSoldTickets'])
        print("價格",b['basePricingInfo']['price'],"報名人數",b['numSoldTickets'],"評價次數",b['numRating'])
print(q)
print(w)
```

```python
import pandas as pd
import matplotlib.pyplot as plt 

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

plt.plot(c,d,'o',color = 'r')
plt.xticks(fontsize=15,rotation=0)
plt.xlabel('評價次數',fontsize=15)
plt.ylabel('報名人數',fontsize=15)
plt.title('hahow分析',fontsize=20)
plt.show()


```
